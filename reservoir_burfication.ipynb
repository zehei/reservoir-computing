```python
import numpy as np
from matplotlib import pyplot as plt
from reservoir import new_network
import three_bit_flip_flop as tbff
```

```python
r = new_network(dim_input=1, dim_reservoir=1000, dim_output=1)
inp, target = tbff.generate_input_and_target(dim_input=r.dim_input, dim_output=r.dim_output, input_duration=100000, num_pulses=1000)
r.train(inp=inp, target=target)
```

```python
#np.save("weight_1000_dim_1", r.weight)
```

```python
inp_test, tar_test = tbff.generate_input_and_target(dim_input=r.dim_input, dim_output=r.dim_output, input_duration=1000, num_pulses=10)
outp_test = r.test(inp_test)
```

```python
xx = np.linspace(1, 1000, 1000)
for i in range(r.dim_output):
    plt.plot(xx, inp_test[:,i])
    plt.plot(xx, outp_test[:,i])
    plt.plot(xx, tar_test[:,i])
    plt.show()
```
